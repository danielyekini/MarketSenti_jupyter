## Import Libraries

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## Fetch Stock Data

Use yfinance to download stock price data. You can specify the stock ticker, the start and end dates, and the interval (e.g., daily, weekly).

```python
aapl_data = yf.download('AAPL', start='2023-01-01', end='2024-01-01', interval='1d')
tsla_data = yf.download('TSLA', start='2023-01-01', end='2024-01-01', interval='1d')
```

## Data Manipulation
With pandas, you can manipulate the data, like calculating moving averages, filtering columns, etc.

```python
aapl_data['Moving Average'] = aapl_data['Close'].rolling(window=50).mean()
tsla_data['Moving Average'] = tsla_data['Close'].rolling(window=50).mean()
```

## Visualisation

For a better understanding of trends, you might want to visualize the data. Libraries like matplotlib or seaborn are great for this.

```python
# Plotting AAPL data
plt.figure(figsize=(10, 5))  # Optional: Adjust the size of the figure
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.plot(aapl_data['Close'], label='AAPL Close Price')
plt.plot(aapl_data['Moving Average'], label='AAPL 50-Day MA')
plt.title('AAPL Stock Price')
plt.legend()

# Plotting TSLA data
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
plt.plot(tsla_data['Close'], label='TSLA Close Price')
plt.plot(tsla_data['Moving Average'], label='TSLA 50-Day MA')
plt.title('TSLA Stock Price')
plt.legend()

plt.tight_layout()  # Adjust the layout so titles and labels don't overlap
plt.show()
```
